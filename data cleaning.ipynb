## Noise Complaints

Data collected from https://data.cityofnewyork.us/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9/data#Filter with the following filter conditions:

* **Complaint type** contains "Noise"
* **Created date** is between Jan 1, 2019 and Jan 1, 2020
* **Borough** is Manhattan

```python
df = pd.read_csv('raw csvs/noise_complaints_raw.csv')
```

```python
df.head()
```

```python
cleaned_df = df.loc[:,["Descriptor", "Latitude", "Longitude"]].dropna()
cleaned_df.head()
```

## Restaurants

Data collected from https://data.cityofnewyork.us/Health/DOHMH-New-York-City-Restaurant-Inspection-Results/43nn-pn8j/data with the following filter conditions:

* **Boro** is Manhattan

```python
df_restos = pd.read_csv('raw csvs/restaurants_raw.csv')
```

```python
df_restos.head()
```

```python
cleaned_df_restos = df_restos.loc[:,["DBA", "CUISINE DESCRIPTION", "Latitude", "Longitude"]].dropna()
cleaned_df_restos.head()
```

```python
cleaned_df_restos.to_csv('exported csvs/restaurants_clean.csv')
```

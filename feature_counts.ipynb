```python
import pandas as pd
```

## Generating comprehensive dataframe with all points

```python
housing = pd.read_csv('manhattan_rentals_apr27.csv')
housing['Latitude'] = housing['Lat_Long'].apply(lambda x: x.replace('[','').replace('\'', '').replace(']','').split(',')[0])
housing['Longitude'] = housing['Lat_Long'].apply(lambda x: x.replace('[','').replace('\'', '').replace(']','').split(',')[1])
housing['Type'] = "Apartment"
housing.head()
```

```python
groceries = pd.read_csv('exported csvs/grocery_stores.csv')
nightlife = pd.read_csv('exported csvs/nightlife_updated.csv')
noise = pd.read_csv('exported csvs/noise_complaints_clean.csv')
restaurants = pd.read_csv('exported csvs/restaurants_clean.csv')
transit = pd.read_csv('exported csvs/subway_entrances.csv')
trees = pd.read_csv('exported csvs/treedata_cleaned.csv')
```

```python
housing_reduced = housing.loc[:,["Listing ID", "Latitude", "Longitude", "Type"]]
housing_reduced.rename(columns={"Listing ID": "Identifier"},inplace=True)
housing_reduced['Latitude'] = housing_reduced['Latitude'].astype(float)
housing_reduced['Longitude'] = housing_reduced['Longitude'].astype(float)
housing_reduced.head()
```

```python
groceries['Type'] = "Grocery"
groceries_reduced = groceries.loc[:,["Entity Name", "Latitude", "Longitude", "Type"]]
groceries_reduced.rename(columns={"Entity Name":"Identifier"},inplace=True)
groceries_reduced.head()
```

```python
nightlife['Type'] = "Nightlife"
nightlife_reduced = nightlife.loc[:,["Premise Name", "lat", "lon", "Type"]]
nightlife_reduced.rename(columns={"Premise Name":"Identifier","lat":"Latitude","lon":"Longitude"},inplace=True)
nightlife_reduced.head()
```

```python
noise['Type'] = "Noise Violation"
noise_reduced = noise.loc[:,["Descriptor", "Latitude", "Longitude", "Type"]]
noise_reduced.rename(columns={"Descriptor":"Identifier"},inplace=True)
noise_reduced.head()
```

```python
restaurants['Type'] = "Restaurant"
restaurants_reduced = restaurants.loc[:, ["DBA", "Latitude", "Longitude", "Type"]]
restaurants_reduced.rename(columns={"DBA":"Identifier"},inplace=True)
restaurants_reduced.head() 
```

```python
transit['Latitude'] = transit['the_geom'].apply(lambda x: x[7:-1].split(' ')[1])
transit['Longitude'] = transit['the_geom'].apply(lambda x: x[7:-1].split(' ')[0])
transit['Type'] = 'Subway'
transit_reduced = transit.loc[:, ["NAME", "Latitude", "Longitude", "Type"]]
transit_reduced.rename(columns={"NAME":"Identifier"},inplace=True)
transit_reduced.head()
```

```python
trees['Type'] = "Tree"
trees_reduced = trees.loc[:, ["tree_id", "latitude", "longitude", "Type"]]
trees_reduced.rename(columns={"tree_id":"Identifier","latitude":"Latitude","longitude":"Longitude"},inplace=True)
trees_reduced.head()
```

```python
#Did not include trees_reduced in an attempt to speed things up.
full_df = pd.concat([groceries_reduced,noise_reduced,restaurants_reduced,transit_reduced,trees_reduced,nightlife_reduced])
full_df['Latitude'] = full_df['Latitude'].astype(float)
full_df['Longitude'] = full_df['Longitude'].astype(float)
```

## Generating geopandas df

```python
import geopandas as gpd

geo_df_features = gpd.GeoDataFrame(
    full_df,
    geometry=gpd.points_from_xy(
        full_df['Longitude'],
        full_df['Latitude'],
    ),
    crs={"init":"EPSG:4326"},)

geo_df_houses = gpd.GeoDataFrame(
    housing_reduced,
    geometry=gpd.points_from_xy(
        housing_reduced['Longitude'],
        housing_reduced['Latitude'],
    ),
    crs={"init":"EPSG:4326"},)

```

```python
features_proj = geo_df_features.to_crs({"init":"EPSG:3857"})
houses_proj = geo_df_houses.to_crs({"init":"EPSG:3857"})
```

## Fill in the number of each feature into the new dataframe

```python
# For each house...
def counts(listing_id,meters=402):
    buffer = houses_proj[houses_proj['Identifier'] == listing_id].buffer(meters).unary_union
    neighbors = features_proj['geometry'].intersection(buffer)
    y = features_proj[~neighbors.is_empty]
    
    num_grocers = len(y[y['Type'] == 'Grocery'])
    num_nightlife = len(y[y['Type'] == 'Nightlife'])
    num_noise = len(y[y['Type'] == 'Noise Violation'])
    num_restaurants = len(y[y['Type'] == 'Restaurant'])
    num_subways = len(y[y['Type'] == 'Subway'])
    num_trees = len(y[y['Type'] == 'Tree'])
    
    return [num_grocers, num_nightlife, num_noise, num_restaurants, num_subways, num_trees]
```

```python
from tqdm import tqdm_notebook

num_f1 = []
num_f2 = []
num_f3 = []
num_f4 = []
num_f5 = []
num_f6 = []

for i in tqdm_notebook(housing['Listing ID']):
    nums = counts(i)
    num_f1.append(nums[0])
    num_f2.append(nums[1])
    num_f3.append(nums[2])
    num_f4.append(nums[3])
    num_f5.append(nums[4])
    num_f6.append(nums[5])
```

```python
import numpy as np

feature_counts = pd.DataFrame(data={
    "num_grocers":num_f1,
    "num_nightlife":num_f2,
    "num_noise":num_f3,
    "num_restaurants":num_f4,
    "num_subways":num_f5,
    "num_trees":num_f6
})

def normalizer(x, column):
    try:
        return (x-np.mean(feature_counts[column]))/np.std(feature_counts[column])
    except:
        return 0
    
feature_counts['grocers_normalized'] = feature_counts['num_grocers'].apply(lambda x: normalizer(x, 'num_grocers'))
feature_counts['nightlife_normalized'] = feature_counts['num_nightlife'].apply(lambda x: normalizer(x, 'num_nightlife'))
feature_counts['noise_normalized'] = feature_counts['num_noise'].apply(lambda x: normalizer(x, 'num_noise'))
feature_counts['restaurants_normalized'] = feature_counts['num_restaurants'].apply(lambda x: normalizer(x, 'num_restaurants'))
feature_counts['subways_normalized'] = feature_counts['num_subways'].apply(lambda x: normalizer(x, 'num_subways'))
feature_counts['trees_normalized'] = feature_counts['num_trees'].apply(lambda x: normalizer(x, 'num_trees'))
```

```python
feature_counts['id'] = housing['Listing ID']
merged = housing.merge(right=feature_counts,left_on='Listing ID',right_on='id')
```

```python
merged.to_csv('full_housing.csv')
```
